```python
import pandas as pd
import re
import seaborn as sns
import os
import matplotlib as plt
import statsmodels.api as sm
```

```python
## Rename columns
df = pd.read_csv(r"C:\Users\Đức\OneDrive - CVA\HUU DUC onedrive\Research paper with Dr.Oleg\dataset\df_match.csv")

df = df.rename(columns={'winner_id': 'p1_id',
                        'winner_seed': 'p1_seed',
                        'winner_name': 'p1_name',
                        'winner_hand': 'p1_hand',
                        'winner_ht': 'p1_ht',
                        'winner_ioc': 'p1_ioc',
                        'winner_age': 'p1_age',
                        'winner_entry': 'p1_entry',
                        'winner_rank': 'p1_rank',
                        'winner_rank_points': 'p1_rank_points'})


df = df.rename(columns={'loser_id': 'p2_id',
                        'loser_seed': 'p2_seed',
                        'loser_name': 'p2_name',
                        'loser_hand': 'p2_hand',
                        'loser_ht': 'p2_ht',
                        'loser_ioc': 'p2_ioc',
                        'loser_age': 'p2_age',
                        'loser_entry': 'p2_entry',
                        'loser_rank': 'p2_rank',
                        'loser_rank_points': 'p2_rank_points'})

df = df.loc[:, df.nunique(dropna=False) > 1]

df['outcome'] = 1

# Add a column to represent difference in seed (there are so many missing values)
df['seed_dif'] = df['p1_seed'] - df['p2_seed']

# Difference in height
df['ht_dif'] = df['p1_ht'] - df['p2_ht']

# Difference in age
df['age_dif'] = df['p1_age'] - df['p2_age']

# Difference in rank and rank points
df['rank_dif'] = df['p1_rank'] - df['p2_rank']
df['rank_points_dif'] = df['p1_rank_points'] - df['p2_rank_points']

# Create dummies variables for surface
df = pd.get_dummies(data = df, columns =['surface'], drop_first=True)
col_list = ['surface_clay', 'surface_grass', 'surface_hard', 'surface_unknown']
for col in col_list:
    df[col] = df[col].apply(lambda x: 0 if x==False else 1)






## Create hand_dif vairable
# df = pd.get_dummies(columns = ['p1_hand', 'p2_hand'], drop_first=True, data = df)

# col_list = ['p1_hand_L', 'p1_hand_R', 'p1_hand_U',
#             'p2_hand_L', 'p2_hand_R', 'p2_hand_U']

# for col in col_list:
#     df[col] = df[col].apply(lambda x: 0 if x==False else 1)
```

```python

X = df[['seed_dif', 'ht_dif', 'age_dif', 'rank_dif', 'rank_points_dif', 'surface_clay', 'surface_grass', 'surface_hard', 'surface_unknown']]
y = df['outcome']

df_model = pd.concat([X,y], axis = 1)
df_model = df_model.dropna()

X = df_model[['seed_dif', 'ht_dif', 'age_dif', 'rank_dif', 'rank_points_dif', 'surface_clay', 'surface_grass', 'surface_hard']]
y = df_model['outcome']
    

X = sm.add_constant(X)
model = sm.Logit(y, X)
result = model.fit()

print(result.summary())

```
